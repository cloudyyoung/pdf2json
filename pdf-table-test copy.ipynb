# PDF Plumber

```python
import pdfplumber
import pandas as pd
import numpy as np

pdf = pdfplumber.open(pdf_path)
```

```python
print(pdf.pages[10].extract_words())
```

```python
print(pdf.pages[10].extract_text())
```

```python
# Turn tables into dataframes

dfs = []
table_settings = {
    "vertical_strategy": "text",
    "horizontal_strategy": "lines",
}

for table in pdf.pages[10].extract_tables(table_settings=table_settings):
    df = pd.DataFrame(table)
    dfs.append(df)
```

```python
im = pdf.pages[10].to_image()
im.reset().debug_tablefinder(table_settings)
```
